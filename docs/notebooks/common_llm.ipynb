# <a name="CLLM"></a>Common LLM Suite - Analysis of Prompts and LLM Outputs
Elemeta has the ability to extract a wide variety of metadata from the text it recieves. The CommonLLMSuite creates one wrapper for the common metadata extractors used for texts. We are also able to use this suite to compare the metadata between prompts and LLM outputs.

```python
!pip install elemeta
!pip install pandas IPython
```

```python
from elemeta.suites.common_llm_suite import CommonLLMSuite
import pandas as pd
from IPython.display import display
cllm = CommonLLMSuite()
md = cllm.run(prompt="What is the purpose of free will?", llm_output="The concept of free will is a complex and philosophical one, and different perspectives exist on its nature and purpose. Generally, free will refers to the ability of individuals to make choices and decisions independently, unconstrained by external forces or predetermined outcomes. The purpose of free will has been debated by philosophers, theologians, and scholars for centuries. Here are a few perspectives on the purpose of free will: Moral Responsibility: One perspective suggests that free will is essential for moral responsibility. If individuals have the ability to make choices, they can be held accountable for their actions. Moral responsibility implies that individuals are morally praised or blamed based on their decisions. Personal Growth and Development: Free will may be seen as a mechanism for personal growth and development. The ability to make choices and learn from experiences allows individuals to develop their character, learn from mistakes, and evolve as individuals. Relationships and Social Order: Free will is often considered crucial for the functioning of social structures. In societies where individuals have the freedom to make choices, there can be diverse perspectives, innovations, and cooperation. The idea is that a society with individuals exercising free will can lead to a more dynamic and adaptive community. Spiritual and Religious Perspectives: In many religious and spiritual traditions, free will is tied to concepts of moral agency and the ability to choose between right and wrong. It may be seen as a gift from a higher power, providing individuals with the opportunity to align themselves with divine will. Existential Fulfillment: From an existentialist perspective, free will is often linked to the concept of existential freedom. The ability to make choices and create one's own meaning in life is considered essential for leading a fulfilling and authentic existence. It's important to note that the philosophical discussions around free will are ongoing, and there are various perspectives on its nature and purpose. Some argue that free will may be limited by factors such as genetics, environment, and neurobiology, raising questions about the extent to which individuals truly have the freedom to choose. The purpose and implications of free will continue to be explored in various fields, including philosophy, psychology, and neuroscience.")
print("Prompt metadata:", md.input_1)
print("LLM output metadata:", md.input_2)
```

Using the CommonLLMSuite on an example prompt and LLM output, we can see some of the typical metadata extracted such as language detected, text complexity, toxicity of text, as well as any similarity of the prompt to jailbreak attempts.
<br>
<br>
Finally, using DataFrame to display the output, we can see the similarity and differences of the values from metadata extractors side-by-side.

```python
prompt_df = pd.DataFrame(list(md.input_1.items()), columns=['Metadata', 'Prompt Value'])
llm_df = pd.DataFrame(list(md.input_2.items()), columns=['Metadata', 'LLM Value'])
result_df = pd.merge(prompt_df, llm_df, on='Metadata', how='outer')
display(result_df)
```
